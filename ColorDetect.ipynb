```python
import cv2

## Read
img = cv2.imread("Pictures/flower.jpg")

## convert to hsv
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

## mask of white
whiteSensitivity = 40
mask1 = cv2.inRange(hsv, (0, 0, 255-whiteSensitivity), (255, whiteSensitivity, 255))

## mask of yellow
yellowSensitivity = 80
mask2 = cv2.inRange(hsv, (20, yellowSensitivity, 255-yellowSensitivity), (45, 255, 255))

## mask of blue (pciture, (lowerRange), (upperRange))
mask3 = cv2.inRange(hsv, (78, 64, 0), (155, 255, 255))

## final mask and masked

#use several masks
#mask = cv2.bitwise_or(mask1, mask2, mask3)

#choose mask for target
mask = mask3

# print mask on target
target = cv2.bitwise_and(img, img, mask=mask)

#save new image with mask
cv2.imwrite("bearbeitet/target.png", target)

#percent of mask
percentage = (mask>0).mean()*100
print(percentage)
```
